# GNN model

````python
from torch.autograd import Variable
# One layer Graph convolution from nodes to edges
class Graph_convolution_v2v_W(nn.Module):
    def __init__(self,in_features,out_features,W,n,bias=True):
        super(Graph_convolution_v2v_W,self).__init__()
        self.register_buffer('w',torch.from_numpy(W.transpose()).float())
        self.register_buffer('n0',torch.tensor(n))
        self.mapping=nn.Parameter(torch.Tensor(n,W.shape[0],W.shape[1]))
        torch.nn.init.xavier_uniform_(self.mapping.data)
        self.scale0=nn.Parameter(torch.Tensor(out_features,in_features))
        torch.nn.init.xavier_uniform_(self.scale0.data)
        self.scale=nn.Parameter(torch.Tensor(n-1,out_features,out_features))
        torch.nn.init.xavier_uniform_(self.scale.data)
        self.bias=nn.Parameter(torch.Tensor(out_features,1))
        torch.nn.init.xavier_uniform_(self.bias.data)

    def forward(self,input):
        h=torch.mul(self.mapping[0,:,:],self.w)
        # print(h.shape,input.shape)
        h=torch.matmul(input,h) # transpose due to batch generator 
        # print(h.shape,self.scale0.shape)
        h=torch.matmul(self.scale0,h)
        for i in range(self.n0 - 1):
          W1 = torch.matrix_power(self.w,i+2)/torch.max(torch.matrix_power(self.w,i+2))
          h1=torch.mul(self.mapping[i,:,:],W1)
          h=torch.matmul(h,h1) # transpose due to batch generator 
          h=torch.matmul(self.scale[i,:,:],h)
        return h+self.bias 

# GNN using DGL v2v graph convolution and our own v2e graph convolution
class GCN(nn.Module):
    def __init__(self, in_feats, hidden_size, W, num_bus, n_hop):
        super(GCN, self).__init__()
        self.conv_v2v1=Graph_convolution_v2v_W(in_feats,hidden_size[0],W,n_hop)
        self.conv_v2v2=Graph_convolution_v2v_W(hidden_size[0],hidden_size[1],W,n_hop)
        self.conv_v2v3=Graph_convolution_v2v_W(hidden_size[1],hidden_size[2],W,n_hop)
        self.conv_v2v4=Graph_convolution_v2v_W(hidden_size[2],hidden_size[3],W,n_hop)
        self.conv_v2v5=Graph_convolution_v2v_W(hidden_size[3],hidden_size[4],W,n_hop)
        self.conv_v2v6=Graph_convolution_v2v_W(hidden_size[4],hidden_size[5],W,n_hop)
        # self.conv_v2v7=Graph_convolution_v2v_W(hidden_size[5],hidden_size[6],W,n_hop)
        # self.conv_v2v8=Graph_convolution_v2v_W(hidden_size[6],hidden_size[7],W,n_hop)
        self.lin_output = nn.Linear(num_bus,num_bus)
         
    def forward(self, inputs):
        # m = nn.ELU()
        # m = nn.LeakyReLU()
        # m = nn.Tanh()
        m = nn.ReLU()
        h=self.conv_v2v1(inputs)
        h=m(h)
        h=self.conv_v2v2(h)
        h=m(h)
        h=self.conv_v2v3(h)
        h=m(h)
        h=self.conv_v2v4(h)
        h=m(h)
        h=self.conv_v2v5(h)
        h=m(h)
        h=self.conv_v2v6(h)
        # h=m(h)
        # h=torch.relu(h)
        # h = self.lin_output(h)
        # time.sleep(5)
        return h
n_bus=x.shape[0]
# w_params=[10,15,10,1]
w_params=[10,15,15,10,10,1]
n_hop = 2 # star shape
# W = W / 10 # normalize
net=GCN(5,w_params,W,n_bus, n_hop) # Laplacian
net=net.to(device)

# set loss_func & optimizer

optimizer=torch.optim.Adam(net.parameters(),weight_decay=0.001)
loss_optm=[]
cvar_optm=[]
cvar_optm_v=[]
cvar_val=[]
cvar_val_v=[]
loss_val=[]
print(net)
````

````python
# v_gnn = np.matmul(R_mat,-p) + np.matmul(X_mat,qg_pred_gnn[1:,:] - q) # lin model
# import torch.nn as nn
# import torch.nn.functional as F
# x # p_c (p), p_g, q_c

# threshold function for dv
class CVaR_loss(nn.Module):
  def __init__(self):
    super(CVaR_loss,self).__init__()
  def forward(self,x,y,thresh,R_mat,X_mat,n,Sphase):
    # print(thresh,thresh.dtype)
    # print(x.dtype,x)
    thresh=torch.tensor(thresh).double()
    x1=x[:,0,1:].double() # p
    x2=x[:,2,1:].double() # q
    y1=y[:,0,1:].double()
    y1 = y1 / Sphase
    x1 = x1 / Sphase
    x2 = x2 / Sphase
    # print(x.shape,y.shape)
    v_dev = torch.matmul(-x1,R_mat) + torch.matmul(y1 - x2,X_mat) # lin model

    # print(v_dev)
    # xx

    # violation = torch.abs(v_dev - thresh)
    violation1=torch.where(v_dev<-thresh,-v_dev,0.)
    violation2=torch.where(v_dev>thresh,v_dev,0.)
    violation = (violation1 + violation2) *n *n

    violation_sum = torch.sum(violation,dim=1)
    # q = torch.quantile(violation,0.9,dim=1,keepdim=True)
    q = torch.quantile(violation,0.8,keepdim=True)
    # print(violation.shape,q.shape,(violation - q).shape)
    output = violation_sum - q
    output = F.softplus(output) # nn.Softmax, F.softmax
    output = output + q
    # print(output,q)
    # xx
    # print('output:',output.shape)
    # print('vdev:',v_dev.shape,'violation:',violation.shape,'quantile:',q.shape)
    # xx
    # output=torch.where(x<-thresh,-1.,x)
    return output #/(n*0.9)

# threshold function for q_g
class CVaR_loss_qg(nn.Module):
  def __init__(self):
    super(CVaR_loss_qg,self).__init__()
  def forward(self,y_true,y_pred):
    # print(thresh,thresh.dtype)
    # print(x.dtype,x)
    violation = torch.linalg.norm(y_true - y_pred,ord=2,dim=2)
    # violation = torch.abs(y_true - y_pred)
    # print(violation.shape)
    # xx

    # violation_sum = torch.sum(violation,dim=1)
    # q = torch.quantile(violation,0.9,dim=1,keepdim=True)
    a = 0.8
    q = torch.quantile(violation,a,keepdim=True)
    # print(violation.shape,q.shape)
    # print(violation.shape,q.shape,(violation - q).shape)
    # output = violation_sum - q
    output = torch.sum(F.softplus(violation - q)) / (100 * a) # nn.Softmax
    output = output + q

    return output #/(n*0.9)    

# set loss_func & optimizer
loss_func=nn.MSELoss()
CVaR_loss_v = CVaR_loss()
CVaR_loss_term = CVaR_loss_qg()
def my_loss_func(pred,label,pred_v,W1,x_batch,thresh,R_mat,X_mat,n,Sphase):
# def my_loss_func(pred,label):
  lmda1=torch.from_numpy(np.array([1])).to(device) # L_2
  lmda2=torch.from_numpy(np.array([0.1])).to(device) # L_inf
  lmda3=torch.from_numpy(np.array([10])).to(device) # CVaR_q
  # lmda4=torch.from_numpy(np.array([1])).to(device) # CVaR_v
  mse=loss_func(pred,label)
  linf=(pred-label).norm(p=float('inf'))
  # cvar_v = CVaR_loss_v(x_batch,pred_v,thresh,R_mat,X_mat,n,Sphase)
  cvar = CVaR_loss_term(label,pred)
  # print(cvar)
  # print(cvar.shape,torch.mean(cvar))
  # return cvar
  # return torch.mean(cvar)
  # return lmda1*mse + lmda3*torch.mean(cvar) # mean/sum over batch
  return lmda1*mse + lmda2*linf + lmda3*torch.mean(cvar) #+ lmda4*torch.mean(cvar_v) # mean/sum over batch
  # return mse
````

````python
max_epochs=200
eval_epoch=5
t0 = time.time()

# earlystopping
tolerance= 5
min_delta=1e-4
previous=0

W_tensor = torch.from_numpy(W).float().to(device)
R_tensor = torch.from_numpy(R_mat).transpose(0,1).double().to(device)
X_tensor = torch.from_numpy(X_mat).transpose(0,1).double().to(device)
dv_max = np.ones(n_node - 1) * 0.03
dv_thresh = torch.from_numpy(dv_max).float().to(device)
n_batch = torch.tensor(params['batch_size']).float().to(device)
sphase = torch.tensor(Sphase).float().to(device)

## main training loop
for epoch in range(max_epochs):
  t10 = time.time()
  # init. loss
  train_loss=0.0
  cvar_loss = 0.0
  cvar_loss_v = 0.0
  batch_num = 0
  batch_trained = 0

# set the list for cvar thres, batch size 200, then 58 batches
# also need to refresh the actual #batch for tracking loss

  if epoch < 5: # first few loops for full training as initialization
    for local_batch0,local_label0 in training_generator:
      batch_num += 1
      # print(batch_num,local_batch0.shape)

      optimizer.zero_grad() # clear the past gradient
      # print(local_batch)
      # print(local_batch0.shape) # [200, 3, 123]
      # print(local_label0.shape) # [200, 1, 123]

  # concatenate for GNN i/o
      # x: p_c,p_g,q_c,p_c_sum,q_c_sum
      # x = np.zeros((n_qg_idx,3+2,params['batch_size'])) # [6,5,batch]
      # y = np.zeros((n_qg_idx,params['batch_size']))
      n_temp = torch.Tensor.size(local_batch0,0) # batch size (last might be different)
      local_batch = torch.zeros((n_temp,5,n_qg_idx)).to(device)
      local_label = torch.zeros((n_temp,1,n_qg_idx)).to(device)

      for i in range(n_qg_idx):
        # x[i,0:3,:] = x0[qg_idx[i],:,:].copy() # p_c,p_g,q_c
        # x[i,3,:] = x0[0,0,:].copy() # p_c_sum
        # x[i,4,:] = x0[0,2,:].copy() # q_c_sum
        # y[i,:] = y0[qg_idx[i],:].copy()
        local_batch[:,0:3,i] = local_batch0[:,:,qg_idx[i]] # p_c,p_g,q_c
        local_batch[:,3,i] = local_batch0[:,0,0] # p_c_sum
        local_batch[:,4,i] = local_batch0[:,2,0] # q_c_sum
        local_label[:,:,i] = local_label0[:,:,qg_idx[i]]


      # GNN
      logits=net(local_batch)
      # print(logits.shape)
      # recover nodal features for loss
      logits0 = torch.zeros((n_temp,1,n_bus)).to(device)
      logits0[:,:,qg_idx] = logits

      # loss
      loss=my_loss_func(logits,local_label,logits0,W_tensor,local_batch0,dv_thresh,R_tensor,X_tensor,n_batch,sphase) #my_loss_func(pred,label,W1,x_batch,thresh,R_mat,X_mat,n)
      # loss=my_loss_func(logits,local_label)
      loss.backward()
      train_loss+=loss.item()
      cvar = CVaR_loss_term(local_label,logits) # CVaR qg
      cvar_loss+=torch.mean(cvar).item() # track CVaR qg loss
      cvar_v = CVaR_loss_v(local_batch0,logits0,dv_thresh,R_tensor,X_tensor,n_batch,sphase) # CVaR dv
      cvar_loss_v+=torch.mean(cvar_v).item() # track CVaR dv loss
      optimizer.step() # update parameters of net
      batch_trained += 1
    loss_optm.append(train_loss/len(training_generator.dataset))
    cvar_optm.append(cvar_loss/len(training_generator.dataset))
    cvar_optm_v.append(cvar_loss_v/len(training_generator.dataset))
    t1 = time.time()
    print("Epoch %d | Time %.2fs  | Training loss: %.4f | CVaR qg: %.4f | CVaR dv: %.4f"%(epoch,t1-t10,train_loss/len(training_generator.dataset),cvar_loss/len(training_generator.dataset),cvar_loss_v/len(training_generator.dataset)))
    
# ------------------------------------------------------------------------------  
# Execute the sampling algorithm approximated by batch selection
# after the first 5 epoches
  else: 
    # Initialize the CVaR list, first 5 batches as ref
    cvar_list = torch.zeros(10)
    for local_batch0,local_label0 in training_generator:
      batch_num += 1
      # print(batch_num,local_batch0.shape)

      optimizer.zero_grad() # clear the past gradient
      # print(local_batch)
      # print(local_batch0.shape) # [200, 3, 123]
      # print(local_label0.shape) # [200, 1, 123]

  # concatenate for GNN i/o
      # x: p_c,p_g,q_c,p_c_sum,q_c_sum
      # x = np.zeros((n_qg_idx,3+2,params['batch_size'])) # [6,5,batch]
      # y = np.zeros((n_qg_idx,params['batch_size']))
      n_temp = torch.Tensor.size(local_batch0,0) # batch size (last might be different, i.e. slightly smaller)
      local_batch = torch.zeros((n_temp,5,n_qg_idx)).to(device)
      local_label = torch.zeros((n_temp,1,n_qg_idx)).to(device)

      for i in range(n_qg_idx):
        # x[i,0:3,:] = x0[qg_idx[i],:,:].copy() # p_c,p_g,q_c
        local_batch[:,0:3,i] = local_batch0[:,:,qg_idx[i]] # p_c,p_g,q_c
        local_batch[:,3,i] = local_batch0[:,0,0] # p_c_sum
        local_batch[:,4,i] = local_batch0[:,2,0] # q_c_sum
        local_label[:,:,i] = local_label0[:,:,qg_idx[i]]


      # GNN
      logits=net(local_batch) # nodal features of nodes with qg support
      # print(logits.shape)

      # recover nodal features for loss
      logits0 = torch.zeros((n_temp,1,n_bus)).to(device)
      logits0[:,:,qg_idx] = logits

      # Calculate CVaR first (weighted sum of both qg and dv)
      cvar = CVaR_loss_term(local_label,logits) # CVaR qg
      # cvar_v = CVaR_loss_v(local_batch0,logits0,dv_thresh,R_tensor,X_tensor,n_batch,sphase) # CVaR dv
      cvar_temp = torch.mean(cvar).item() #+ torch.mean(cvar_v).item()

      # Determine if the current batch is kept
      if batch_num < 6: # first 5
        # print(batch_num)
        cvar_list[batch_num-1] = cvar_temp
        # execute training for first 5 batches
        batch_trained += 1
        # loss
        loss=my_loss_func(logits,local_label,logits0,W_tensor,local_batch0,dv_thresh,R_tensor,X_tensor,n_batch,sphase) #my_loss_func(pred,label,W1,x_batch,thresh,R_mat,X_mat,n)
        # loss=my_loss_func(logits,local_label)
        loss.backward()
        train_loss+=loss.item()
        
        cvar_loss+=torch.mean(cvar).item() # track CVaR qg loss
        
        cvar_loss_v+=torch.mean(cvar_v).item() # track CVaR dv loss
        optimizer.step() # update parameters of net
      else:
        cvar_thres = torch.quantile(cvar_list,0.5)
        if cvar_temp > cvar_thres:
          # execute training
          batch_trained += 1
          # loss
          loss=my_loss_func(logits,local_label,logits0,W_tensor,local_batch0,dv_thresh,R_tensor,X_tensor,n_batch,sphase) #my_loss_func(pred,label,W1,x_batch,thresh,R_mat,X_mat,n)
          # loss=my_loss_func(logits,local_label)
          loss.backward()
          train_loss+=loss.item()
          
          cvar_loss+=torch.mean(cvar).item() # track CVaR qg loss
          
          cvar_loss_v+=torch.mean(cvar_v).item() # track CVaR dv loss
          optimizer.step() # update parameters of net
        
        # Update the CVaR list
        min0 = torch.min(cvar_list,dim=0,keepdim=False, out=None)
        if cvar_temp > min0[0]:
          cvar_list[min0[1]] = cvar_temp


    print('Epoch %d | Trained batch: %d'%(epoch,batch_trained)) 
    loss_optm.append(train_loss/(len(training_generator.dataset)*(batch_trained/58)))
    cvar_optm.append(cvar_loss/(len(training_generator.dataset)*(batch_trained/58)))
    cvar_optm_v.append(cvar_loss_v/(len(training_generator.dataset)*(batch_trained/58)))
    t1 = time.time()
    print("Epoch %d | Time %.2fs  | Training loss: %.4f | CVaR qg: %.4f | CVaR dv: %.4f"%(epoch,t1-t10,train_loss/len(training_generator.dataset),cvar_loss/len(training_generator.dataset),cvar_loss_v/len(training_generator.dataset)))
# Evaluation epoch
  if (epoch+1)%eval_epoch==0:
    net.eval()
    eval_loss=0.0
    cvar_eval_loss=0.0
    cvar_eval_loss_v=0.0
    for eval_batch,eval_label in validation_generator:
      eval_batch,eval_label=eval_batch.to(device),eval_label.to(device)
      n_temp1 = torch.Tensor.size(eval_batch,0) # batch size (last might be different)

      # concatenate for GNN i/o 
      eval_batch1 = torch.zeros((n_temp1,5,n_qg_idx)).to(device)
      eval_label1 = torch.zeros((n_temp1,1,n_qg_idx)).to(device)

      for i in range(n_qg_idx):
        eval_batch1[:,0:3,i] = eval_batch[:,:,qg_idx[i]] # p_c,p_g,q_c
        eval_batch1[:,3,i] = eval_batch[:,0,0] # p_c_sum
        eval_batch1[:,4,i] = eval_batch[:,2,0] # q_c_sum
        eval_label1[:,:,i] = eval_label[:,:,qg_idx[i]]

      logits=net(eval_batch1)
      # recover nodal features for loss
      logits1 = torch.zeros((n_temp1,1,n_bus)).to(device)
      logits1[:,:,qg_idx] = logits

      loss=my_loss_func(logits,eval_label1,logits1,W_tensor,eval_batch,dv_thresh,R_tensor,X_tensor,n_batch,sphase)
      cvar = CVaR_loss_term(eval_label1,logits) # CVaR
      cvar_v = CVaR_loss_v(eval_batch,logits1,dv_thresh,R_tensor,X_tensor,n_batch,sphase) # CVaR dv
      eval_loss+=loss.item()
      cvar_eval_loss+=torch.mean(cvar).item()
      cvar_eval_loss_v+=torch.mean(cvar_v).item()
    
    eval_avg=eval_loss/len(validation_generator.dataset)
    cvar_eval_avg=cvar_eval_loss/len(validation_generator.dataset)
    cvar_eval_avg_v=cvar_eval_loss_v/len(validation_generator.dataset)
    if (epoch==0): previous=eval_avg
    else:
      if previous-eval_avg<min_delta: tolerance-=1
      if tolerance==0: break
      previous=eval_avg
    print('Evaluation:')
    print("Epoch %d | Eval loss: %.4f | CVaR qg eval: %.4f | CVaR dv eval: %.4f" % (epoch, eval_avg, cvar_eval_avg, cvar_eval_avg_v))
    loss_val.append([epoch, eval_loss/len(validation_generator.dataset)])
    cvar_val.append([epoch, cvar_eval_loss/len(validation_generator.dataset)])
    cvar_val_v.append([epoch, cvar_eval_loss_v/len(validation_generator.dataset)])
    net.train()
  # validation loop
  with torch.set_grad_enabled(False):
      pass

t2 = time.time()
print("Training time:%.2fs"%(t2-t0))
````
